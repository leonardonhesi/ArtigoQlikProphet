```python
import pandas              as pd
import numpy               as np
import plotly.offline      as py
import joblib              as jb
from fbprophet             import Prophet
from fbprophet.plot        import plot_plotly
```

```python
df = pd.read_excel('./QlikProphet/DataSetATUAL.xlsx')
df.tail(20)
```

```python
df['ValorAtual']        = df['ValorAtual'].apply(lambda x: np.nan if x == '-' else x)
df                      = df.dropna()
df['Data']              = pd.to_datetime(df['Data'])
df                      = df.filter(items=['Data', 'ValorAtual'])
df                      = df.set_index('Data').sort_index()
df                      = df.reset_index(drop=False)
df.columns              = ['ds', 'y']
df.tail(20)
```

```python
# input_qlik
# future_qlik
prophet2            = Prophet(changepoint_prior_scale=0.5)
prophet2.add_country_holidays(country_name='BR')
df2                 = df[['ds','y']].dropna()
# fit(df, algorithm='Newton')
prophet2.fit(df2, algorithm='LBFGS')
futuro2             = prophet2.make_future_dataframe(periods=8, freq="D")
forecast2           = prophet2.predict(futuro2)
```

```python
forecast2[['ds','yhat']].tail(20)
```

```python
pred_df2            = forecast2.set_index('ds')[['yhat', 'yhat_lower', 'yhat_upper']].join(df2.set_index('ds').y).reset_index()
pred_df2.tail(20)
```

```python
fig6                = plot_plotly(prophet2, forecast2)
py.iplot(fig6)
```
