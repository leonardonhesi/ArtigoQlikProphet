```python
import pandas as pd
```

```python
df = pd.read_csv('2019-1_CA.csv', delimiter=';')
df.head()
```

```python
df = df[df['Estado - Sigla'] == 'SP']
df = df[df['Município']      == 'ITU']
df = df[df['Produto']        == 'GASOLINA']
df.head()
```

```python
df = df.filter(items=['Bandeira','Revenda','CNPJ da Revenda', 'Data da Coleta', 'Valor de Venda'])
df.head()
```

```python
df['Data da Coleta']    = pd.to_datetime(df['Data da Coleta'])
df                      = df.set_index('Data da Coleta').sort_index()
df                      = df.reset_index(drop=False)
df.head()
```

```python

df['Data da Coleta'].describe()
```

```python
# first:2019-01-14 last:2019-12-06
time_index = pd.date_range(start='14/01/2019', end='06/12/2019', freq='D')
dfTS = pd.DataFrame(index=time_index)
#df['Sales'] = [1.0,2.0,np.nan,np.nan,5.0]
#df.ffill()
dfTS                      = dfTS.reset_index(drop=False)
dfTS
```

```python
result = dfTS.set_index('index').join(df.set_index('Data da Coleta'))
result.head(20)
```
